```
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

##### Wav2Vec2ForCTC: This is a model from the Wav2Vec 2.0 family, designed for automatic speech recognition (ASR). "CTC" stands for "Connectionist Temporal Classification," which helps map sequences of audio frames to text sequences, even if they have different lengths.

##### Wav2Vec2Tokenizer: This component tokenizes audio input into a format that the Wav2Vec2 model can process. It converts raw audio data into tokens that the model understands and later decodes them back into human-readable text.

```
#load pre-trained model and tokenizer
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```
#load any audio file of your choice
speech, rate = librosa.load("quote.wav",sr=16000)
```

```
import IPython.display as display
display.Audio("quote.wav", autoplay=True)
```

```
input_values = tokenizer(speech, return_tensors = 'pt').input_values
```

```
input_values
```

```
#Store logits (non-normalized predictions)
logits = model(input_values).logits
```

```
logits
```

```
#Store predicted id's
predicted_ids = torch.argmax(logits, dim =-1)
```

```
#decode the audio to generate text
transcriptions = tokenizer.decode(predicted_ids[0])
```

```
print(transcriptions)
```
